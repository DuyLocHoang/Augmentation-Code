```python
import cv2
import os
import albumentations as alb
import glob
```

```python
root_path = "ts\\ts\\"

for file in glob.glob(root_path + '*.txt'):
    bbox_images = []
    labels_names = []
    with open(file,'r') as f:
        data = f.readlines()
        for i in data :
            convert = i.split()
            h = [float(i) for i in convert[1:]]
#             print(h)
            bbox_images.append(h)
            labels_names.append(convert[0])
    print(bbox_images)
    print(labels_names)
print('\n')
```
